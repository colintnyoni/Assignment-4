```python
#Pie Chart Representation of Customers, gender based.
#
Gender= ['Male','Female']
slices = [3555,3488]
cols = ['b','r']
plt.pie(slices,
        labels=Gender,
        colors=cols,
        startangle=90,
        shadow= True,
        explode=(0,0.1),
        autopct='%1.1f%%')

plt.title('Ratio of Customers Gender-based')
plt.show()
```
